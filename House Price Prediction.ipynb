## Data Cleaning

### For columns "Location"

```python
# Removing space of locations
data["location"] = data["location"].apply(lambda x : x.strip())
```

```python
location_stats = data.groupby("location")["location"].agg("count").sort_values(ascending = False)
location_stats
```

```python
location_less_than_10_entries = location_stats[location_stats <= 10]
location_less_than_10_entries
```

```python
# entries that have less than 10 is converted into "other"
data["location"] = data["location"].apply(lambda x:"other" if x in location_less_than_10_entries else x)
```

```python
data["location"].value_counts()
```

### For column "Size"

```python
data["size"].unique()
```

```python
data["bedrooms"] = data["size"].apply(lambda x:int(x.split(" ")[0]))
data
```

### For column "Total sqft"

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    tokens = sqft.split("-")
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
```

```python
data["total_sqft"] = data["total_sqft"].apply(clean)
data.total_sqft.unique()
```

```python
data.dropna(inplace = True)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

sparse_output=False: By default, OneHotEncoder outputs a sparse matrix, which is a memory-efficient way to store data that is mostly zeros. Setting this to False tells the encoder to output a regular, dense NumPy array, which is often easier to look at and work with directly.

```python
ohe
```

```python
# One-hot encode the 'location' column and keep all other columns.
# Create a preprocessor to apply one-hot encoding to the 'location' column, passing through the rest.
col_trans = make_column_transformer((ohe, ["location"]), remainder="passthrough")
```

(ohe, ["location"]): This is a tuple that acts as an instruction. It tells the ColumnTransformer: "Take the ohe object you created earlier and apply it only to the column named "location"."

remainder="passthrough": This is a crucial argument. It answers the question, "What should I do with all the other columns that you didn't tell me to transform (i.e., every column except "location")?" The value "passthrough" means: "Just leave them as they are and carry them over to the final output without any changes." The other common option is "drop", which would discard them entirely.

```python
lr = LinearRegression()
ss = StandardScaler()
```

```python
model = make_pipeline(col_trans, ss, lr)
```

```python
#spliting data into input and output variable
x = data.drop(columns = "price") # Input
y = data["price"] # Output
```

```python
# Spliting in training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)*100
```

```python
model.score(x_test, y_test)*100
```

```python
input = pd.DataFrame([["Electronic City Phase II", 2000.0, 3.0, 2.0, 3]], columns = ["location", "total_sqft", "bath","balcony", "bedrooms"])
```
